```python
def resolve_dotted_name(name):
    parts = name.split('.')
    obj = importlib.import_module(parts[0])
    for part in parts[1:]:
        obj = getattr(obj, part)
    return obj

groups = manifest['groups']
assert len(groups) == manifest['group_count'] == 10
assert all(len(group['apis']) == manifest['apis_per_group'] == 12 for group in groups)
apis = [
    {'group': group['id'], 'api': api}
    for group in groups
    for api in group['apis']
]
assert len(apis) == manifest['target_api_count'] == 120
assert len({entry['api'] for entry in apis}) == 120

missing, non_callable = [], []
for entry in apis:
    try:
        obj = resolve_dotted_name(entry['api'])
    except (AttributeError, ImportError) as exc:
        missing.append((entry['api'], str(exc)))
        continue
    if not callable(obj):
        non_callable.append(entry['api'])
assert not missing, missing
assert not non_callable, non_callable

seeds = manifest['generation_seeds']
baseline_ids = manifest['baseline_ids']
assert len(seeds) == len(set(seeds)) == 5
assert baseline_ids == ['B0', 'B1', 'B2', 'B3']

tasks = []
position_counts = Counter()
for seed_index, seed in enumerate(seeds):
    for api_index, entry in enumerate(apis):
        rotation = (api_index + seed_index) % len(baseline_ids)
        baseline_order = baseline_ids[rotation:] + baseline_ids[:rotation]
        for position, baseline_id in enumerate(baseline_order):
            position_counts[(baseline_id, position)] += 1
        tasks.append({
            'task_id': f"{manifest['benchmark_id']}|{entry['api']}|{seed}",
            'api': entry['api'],
            'group': entry['group'],
            'api_index': api_index,
            'seed_index': seed_index,
            'generation_seed': seed,
            'baseline_order': baseline_order,
            'ab_order': ('B2_then_B3' if (seed_index < 4 and seed_index % 2 == 0) or (seed_index == 4 and api_index < 60) else 'B3_then_B2'),
        })

assert len(tasks) == 600
assert set(position_counts.values()) == {150}, position_counts
assert Counter(task['ab_order'] for task in tasks) == {'B2_then_B3': 300, 'B3_then_B2': 300}
print('Runtime API validation: PASS')
print('Task matrix: PASS — 600 paired tasks / 2,400 baseline executions')
print('Logical Latin schedule: PASS — every baseline 150x in every recorded position')
print('Physical B2/B3 schedule: PASS — 300 B2-first and 300 B3-first tasks')
```

## Fair A/B contract

- B2 and B3 use identical API–seed task IDs, full prompt bytes, documentation snapshot, decoding parameters, token budget, retries, timeout, validator, subprocess and oracle.
- Prefer one frozen base checkpoint with adapter **off** for B2 and **on** for B3.
- The generation seed must be passed to the model sampler and host/framework RNGs.
- Missing APIs fail closed; the notebook never silently substitutes another API.
- Results are paired by task_id, not compared as unrelated aggregate runs.

```python
import ast, importlib.metadata, inspect, os, platform, re, subprocess, sys, tempfile, time
from datetime import datetime, timezone

EVENTS_PATH = OUTPUT_DIR / 'events.jsonl'
RUN_MANIFEST_PATH = OUTPUT_DIR / 'run_manifest.json'
DOCS_PATH = OUTPUT_DIR / 'documentation_snapshot.json'

def utc_now():
    return datetime.now(timezone.utc).isoformat()

def sha256_text(text):
    return hashlib.sha256(text.encode('utf-8')).hexdigest()

def package_version(name):
    try:
        return importlib.metadata.version(name)
    except importlib.metadata.PackageNotFoundError:
        return 'not-installed'

def safe_doc_and_signature(api):
    obj = resolve_dotted_name(api)
    try:
        signature = str(inspect.signature(obj))
    except Exception:
        signature = '(signature unavailable)'
    try:
        doc = inspect.getdoc(obj) or ''
    except Exception:
        doc = ''
    doc = re.sub(r'\n{3,}', '\n\n', doc).strip()[:2400]
    return {'signature': signature[:500], 'doc': doc}

documentation_snapshot = {entry['api']: safe_doc_and_signature(entry['api']) for entry in apis}
documentation_json = json.dumps(documentation_snapshot, ensure_ascii=False, sort_keys=True, indent=2)
documentation_sha256 = sha256_text(documentation_json)
DOCS_PATH.write_text(documentation_json, encoding='utf-8')

FULL_SYSTEM_PROMPT = (
    'You generate rigorous PyTorch API tests. Output exactly one complete Python program and no markdown.'
)
MINIMAL_SYSTEM_PROMPT = 'Output exactly one Python program and no explanation.'

def minimal_messages(api):
    return [
        {'role': 'system', 'content': MINIMAL_SYSTEM_PROMPT},
        {'role': 'user', 'content': f'Write a small executable Python program that calls {api}. Output code only.'},
    ]

def full_messages(api):
    info = documentation_snapshot[api]
    prompt = f'''Target API: {api}
Signature: {info['signature']}
Documentation snapshot:
{info['doc']}

Create a tiny deterministic CPU test for this exact API.
Requirements:
1. Start with import torch and call the exact target API.
2. Use small valid inputs and finish within one minute.
3. Include a meaningful non-tautological oracle when possible: a reference value, metamorphic relation, CPU/eager relation, gradient property, or documented invariant.
4. Do not compare a value with itself and do not assert a constant.
5. Do not use files, network, subprocesses, ctypes, multiprocessing, or exception suppression.
6. Let unexpected exceptions escape.
7. Output only Python code without markdown fences.'''
    return [
        {'role': 'system', 'content': FULL_SYSTEM_PROMPT},
        {'role': 'user', 'content': prompt},
    ]

def prompt_bytes(baseline, api):
    messages = minimal_messages(api) if baseline == 'B1' else full_messages(api)
    return json.dumps(messages, ensure_ascii=False, sort_keys=True, separators=(',', ':'))

def extract_python(text):
    text = (text or '').strip()
    fenced = re.search(r'~~~(?:python)?\s*(.*?)~~~', text, flags=re.I | re.S)
    if not fenced:
        fenced = re.search(r'\x60\x60\x60(?:python)?\s*(.*?)\x60\x60\x60', text, flags=re.I | re.S)
    if fenced:
        text = fenced.group(1).strip()
    start = text.find('import ')
    if start > 0:
        text = text[start:]
    return text

def dotted_name(node):
    if isinstance(node, ast.Name):
        return node.id
    if isinstance(node, ast.Attribute):
        left = dotted_name(node.value)
        return f'{left}.{node.attr}' if left else node.attr
    return ''

ALLOWED_IMPORT_ROOTS = {'torch', 'math', 'numpy', 'typing'}
BANNED_CALL_ROOTS = {'open', 'exec', 'eval', '__import__', 'input', 'breakpoint', 'os', 'sys', 'subprocess', 'socket', 'requests', 'ctypes', 'multiprocessing'}

def inspect_program(code, target_api):
    result = {'parseable': False, 'ast_pass': False, 'target_call_present': False, 'oracle_present': False, 'fake_assertion': False, 'ast_reason': ''}
    try:
        tree = ast.parse(code)
    except SyntaxError as exc:
        result['ast_reason'] = f'syntax: {exc}'
        return result, None
    result['parseable'] = True
    violations = []
    call_names = []
    meaningful_oracle = False
    fake_assertion = False
    for node in ast.walk(tree):
        if isinstance(node, ast.Import):
            for alias in node.names:
                if alias.name.split('.')[0] not in ALLOWED_IMPORT_ROOTS:
                    violations.append(f'import {alias.name}')
        elif isinstance(node, ast.ImportFrom):
            root = (node.module or '').split('.')[0]
            if root not in ALLOWED_IMPORT_ROOTS:
                violations.append(f'from {node.module} import ...')
        elif isinstance(node, ast.Call):
            name = dotted_name(node.func)
            call_names.append(name)
            root = name.split('.')[0]
            if root in BANNED_CALL_ROOTS:
                violations.append(f'call {name}')
            if name == 'torch.testing.assert_close':
                meaningful_oracle = True
        elif isinstance(node, ast.Assert):
            test = node.test
            is_fake = isinstance(test, ast.Constant)
            if isinstance(test, ast.Compare) and len(test.comparators) == 1:
                is_fake = is_fake or ast.dump(test.left) == ast.dump(test.comparators[0])
            for child in ast.walk(test):
                if isinstance(child, ast.Call) and dotted_name(child.func).endswith(('allclose', 'isclose')) and len(child.args) >= 2:
                    is_fake = is_fake or ast.dump(child.args[0]) == ast.dump(child.args[1])
            fake_assertion = fake_assertion or is_fake
            meaningful_oracle = meaningful_oracle or not is_fake
    if len(code.encode('utf-8')) > 20000:
        violations.append('program exceeds 20 KB')
    result['ast_pass'] = not violations
    result['target_call_present'] = target_api in call_names
    result['oracle_present'] = meaningful_oracle
    result['fake_assertion'] = fake_assertion
    result['ast_reason'] = '; '.join(violations)
    return result, tree

def execute_program(code, seed):
    path = None
    started = time.time()
    try:
        with tempfile.NamedTemporaryFile('w', suffix='.py', encoding='utf-8', delete=False) as handle:
            handle.write(code)
            path = handle.name
        env = os.environ.copy()
        env['PYTHONHASHSEED'] = str(seed)
        if FORCE_GENERATED_PROGRAMS_TO_CPU:
            env['CUDA_VISIBLE_DEVICES'] = ''
        proc = subprocess.run([sys.executable, path], capture_output=True, text=True, timeout=SUBPROCESS_TIMEOUT_S, env=env)
        return {'exit_code': proc.returncode, 'stdout': proc.stdout[-4000:], 'stderr': proc.stderr[-4000:], 'timeout': False, 'subprocess_seconds': round(time.time() - started, 3)}
    except subprocess.TimeoutExpired as exc:
        return {'exit_code': None, 'stdout': (exc.stdout or '')[-4000:] if isinstance(exc.stdout, str) else '', 'stderr': (exc.stderr or '')[-4000:] if isinstance(exc.stderr, str) else '', 'timeout': True, 'subprocess_seconds': round(time.time() - started, 3)}
    finally:
        if path:
            try:
                os.unlink(path)
            except OSError:
                pass

def evaluate_code(code, task):
    static, _tree = inspect_program(code, task['api'])
    execution = {'exit_code': None, 'stdout': '', 'stderr': '', 'timeout': False, 'subprocess_seconds': 0.0}
    if static['parseable'] and static['ast_pass']:
        execution = execute_program(code, task['generation_seed'])
    runnable = execution['exit_code'] == 0 and not execution['timeout']
    target_valid = runnable and static['target_call_present']
    oracle_bearing = target_valid and static['oracle_present'] and not static['fake_assertion']
    labels = []
    if not static['parseable']: labels.append('syntax')
    if static['parseable'] and not static['ast_pass']: labels.append('ast_policy')
    if execution['timeout']: labels.append('timeout')
    if static['parseable'] and static['ast_pass'] and not runnable and not execution['timeout']: labels.append('runtime')
    if runnable and not static['target_call_present']: labels.append('wrong_api')
    if target_valid and not static['oracle_present']: labels.append('no_oracle')
    if static['fake_assertion']: labels.append('fake_assertion')
    return {**static, **execution, 'runnable': runnable, 'target_valid': target_valid, 'oracle_bearing': oracle_bearing, 'error_labels': labels}

run_manifest = {
    'created_utc': utc_now(),
    'benchmark_id': manifest['benchmark_id'],
    'manifest_sha256': manifest_sha256,
    'documentation_sha256': documentation_sha256,
    'torch_version': torch.__version__,
    'torch_cuda': torch.version.cuda,
    'python': platform.python_version(),
    'packages': {name: package_version(name) for name in ['llama-cpp-python', 'huggingface-hub', 'numpy', 'pandas']},
    'gpu': torch.cuda.get_device_name(0) if torch.cuda.is_available() else None,
    'base_model': BASE_MODEL,
    'tuned_model': TUNED_MODEL,
    'decoding': {'temperature': TEMPERATURE, 'top_p': TOP_P, 'top_k': TOP_K, 'min_p': MIN_P, 'repeat_penalty': REPEAT_PENALTY, 'max_tokens': MAX_TOKENS},
    'subprocess_timeout_s': SUBPROCESS_TIMEOUT_S,
    'selected_task_count': len(selected_tasks),
}
RUN_MANIFEST_PATH.write_text(json.dumps(run_manifest, ensure_ascii=False, indent=2), encoding='utf-8')
print('Documentation SHA-256:', documentation_sha256)
```

```python
llm = None
loaded_model_label = None

def unload_llm():
    global llm, loaded_model_label
    if llm is not None:
        try:
            llm.close()
        except Exception:
            pass
    llm = None
    loaded_model_label = None
    gc.collect()
    if torch.cuda.is_available():
        torch.cuda.empty_cache()

def verified_staged_model(spec):
    print(f'Downloading/checking pinned artifact: {spec["repo_id"]}/{spec["filename"]}', flush=True)
    cached_path = Path(hf_hub_download(repo_id=spec['repo_id'], filename=spec['filename'], revision=spec['revision'], token=HF_TOKEN, cache_dir=str(MODEL_CACHE_DIR)))
    if cached_path.stat().st_size != spec['file_size']:
        raise RuntimeError(f'Artifact size mismatch for {cached_path}')
    verified_marker = cached_path.with_name(cached_path.name + f'.{spec["file_sha256"]}.verified')
    if not verified_marker.exists():
        print('Verifying SHA-256 once; marker will persist on Drive...', flush=True)
        digest = hashlib.sha256()
        with cached_path.open('rb') as stream:
            while chunk := stream.read(16 * 1024 * 1024):
                digest.update(chunk)
        if digest.hexdigest() != spec['file_sha256']:
            raise RuntimeError(f'SHA-256 mismatch for {cached_path}')
        verified_marker.write_text(spec['file_sha256'] + '\n', encoding='ascii')
    if not IS_COLAB:
        return str(cached_path)
    local_dir = Path('/content/trdgl_model_stage')
    local_dir.mkdir(parents=True, exist_ok=True)
    local_path = local_dir / spec['filename']
    if not local_path.exists() or local_path.stat().st_size != spec['file_size']:
        partial_path = local_path.with_suffix(local_path.suffix + '.partial')
        if partial_path.exists():
            partial_path.unlink()
        required_free = spec['file_size'] + 2 * 2**30
        if shutil.disk_usage(local_dir).free < required_free:
            print('Local SSD is tight; removing older staged GGUF copies (persistent source cache is kept)...', flush=True)
            for old_stage in local_dir.glob('*.gguf'):
                if old_stage != local_path:
                    old_stage.unlink()
            for stale_partial in local_dir.glob('*.partial'):
                if stale_partial != partial_path:
                    stale_partial.unlink()
        if shutil.disk_usage(local_dir).free < required_free:
            raise RuntimeError(f'Insufficient local SSD space to stage {spec["filename"]}: need {required_free / 2**30:.1f} GiB free')
        print(f'Staging {spec["file_size"] / 2**30:.2f} GiB from Drive to local SSD...', flush=True)
        copied = 0
        next_report = 2 * 2**30
        with cached_path.open('rb') as source, partial_path.open('wb') as target:
            while chunk := source.read(16 * 1024 * 1024):
                target.write(chunk)
                copied += len(chunk)
                if copied >= next_report:
                    print(f'  staged {copied / 2**30:.1f} GiB', flush=True)
                    next_report += 2 * 2**30
        partial_path.replace(local_path)
    return str(local_path)

def load_llm(label, spec):
    global llm, loaded_model_label
    if loaded_model_label == label and llm is not None:
        return
    unload_llm()
    print(f'Preparing {label}: {spec["repo_id"]}@{spec["revision"]}', flush=True)
    model_path = verified_staged_model(spec)
    attempts = [(N_CTX, N_GPU_LAYERS), (2048, 60), (2048, 48), (1536, 40), (1024, 32)]
    errors = []
    for n_ctx, n_gpu_layers in attempts:
        try:
            llm = Llama(model_path=model_path, n_ctx=n_ctx, n_gpu_layers=n_gpu_layers, seed=manifest['generation_seeds'][0], verbose=False)
            loaded_model_label = label
            print(f'Loaded {label}: n_ctx={n_ctx}, n_gpu_layers={n_gpu_layers}')
            return
        except Exception as exc:
            errors.append(f'n_ctx={n_ctx}, layers={n_gpu_layers}: {exc}')
            unload_llm()
    raise RuntimeError('Model load failed: ' + ' | '.join(errors[-5:]))

def set_true_generation_seed(seed):
    random.seed(seed)
    np.random.seed(seed % (2**32 - 1))
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)
    if llm is None or not hasattr(llm, 'set_seed'):
        raise RuntimeError('This llama.cpp binding does not expose set_seed; refusing an unseeded run.')
    llm.set_seed(seed)

def manual_gemma_prompt(messages):
    chunks = []
    for message in messages:
        role = 'model' if message['role'] == 'assistant' else message['role']
        chunks.append(f'<start_of_turn>{role}\n{message["content"]}<end_of_turn>')
    chunks.append('<start_of_turn>model\n')
    return '\n'.join(chunks)

def generate_seeded(messages, seed):
    set_true_generation_seed(seed)
    common = dict(max_tokens=MAX_TOKENS, temperature=TEMPERATURE, top_p=TOP_P, top_k=TOP_K, min_p=MIN_P, repeat_penalty=REPEAT_PENALTY, seed=seed)
    started = time.time()
    try:
        response = llm.create_chat_completion(messages=messages, **common)
        raw = response['choices'][0]['message']['content']
        finish_reason = response['choices'][0].get('finish_reason')
    except (ValueError, RuntimeError) as chat_exc:
        response = llm(manual_gemma_prompt(messages), stop=['<end_of_turn>', '<start_of_turn>'], echo=False, **common)
        raw = response['choices'][0]['text']
        finish_reason = response['choices'][0].get('finish_reason')
    elapsed = round(time.time() - started, 3)
    token_count = None
    try:
        token_count = len(llm.tokenize(raw.encode('utf-8'), add_bos=False))
    except Exception:
        pass
    return raw, {'generation_seconds': elapsed, 'raw_token_count': token_count, 'finish_reason': finish_reason, 'seed_backend': 'host+torch+cuda+llama_cpp.set_seed+completion(seed)'}
```

```python
RUNNER_VERSION = 'four_baseline_runner_v1'
run_signature_payload = {k: v for k, v in run_manifest.items() if k != 'created_utc'}
run_signature_payload.update({'runner_version': RUNNER_VERSION, 'minimal_system_prompt': MINIMAL_SYSTEM_PROMPT, 'full_system_prompt': FULL_SYSTEM_PROMPT, 'selected_task_ids': [task['task_id'] for task in selected_tasks], 'run_baselines': RUN_BASELINES})
run_signature = sha256_text(json.dumps(run_signature_payload, ensure_ascii=False, sort_keys=True, separators=(',', ':')))
run_manifest['run_signature'] = run_signature
run_manifest['event_log'] = str(EVENTS_PATH)
RUN_MANIFEST_PATH.write_text(json.dumps(run_manifest, ensure_ascii=False, indent=2), encoding='utf-8')

def read_completed():
    done = set()
    if not EVENTS_PATH.exists():
        return done
    with EVENTS_PATH.open('r', encoding='utf-8') as handle:
        for line in handle:
            try:
                event = json.loads(line)
            except json.JSONDecodeError:
                continue
            if event.get('run_signature') == run_signature:
                done.add((event['baseline'], event['task_id']))
    return done

completed = read_completed()
print('Resume: completed events for this run signature:', len(completed))

def append_event(event):
    with EVENTS_PATH.open('a', encoding='utf-8') as handle:
        handle.write(json.dumps(event, ensure_ascii=False) + '\n')
        handle.flush()
        os.fsync(handle.fileno())
    completed.add((event['baseline'], event['task_id']))

def run_one(baseline, task):
    key = (baseline, task['task_id'])
    if key in completed or baseline not in RUN_BASELINES:
        return
    started_utc = utc_now()
    if baseline == 'B0':
        raw = b0_program(task['api'], task['generation_seed'])
        generation_meta = {'generation_seconds': 0.0, 'raw_token_count': None, 'finish_reason': 'template', 'seed_backend': 'host+torch template'}
        prompt_serialized = 'B0_TYPED_TEMPLATE_V1'
        model = 'typed_random_template_v1'
    else:
        messages = minimal_messages(task['api']) if baseline == 'B1' else full_messages(task['api'])
        prompt_serialized = json.dumps(messages, ensure_ascii=False, sort_keys=True, separators=(',', ':'))
        raw, generation_meta = generate_seeded(messages, task['generation_seed'])
        model = loaded_model_label
    code = extract_python(raw)
    measured = evaluate_code(code, task)
    event = {
        'run_signature': run_signature,
        'started_utc': started_utc,
        'finished_utc': utc_now(),
        'baseline': baseline,
        'model': model,
        'task_id': task['task_id'],
        'api': task['api'],
        'api_group': task['group'],
        'api_index': task['api_index'],
        'generation_seed': task['generation_seed'],
        'ab_order': task['ab_order'],
        'logical_baseline_order': task['baseline_order'],
        'prompt_sha256': sha256_text(prompt_serialized),
        'raw_output_sha256': sha256_text(raw),
        'raw_generation': True,
        'raw_output': raw,
        'extracted_code': code,
        **generation_meta,
        **measured,
    }
    append_event(event)
    print(f"[{len(completed):4d}] {baseline} seed={task['generation_seed']} {task['api']} -> runnable={event['runnable']} target={event['target_valid']} oracle={event['oracle_bearing']}")

# B0 is independent of model loading and is checkpointed first.
for task in selected_tasks:
    run_one('B0', task)

# Exact A/B order balance: 300 tasks B2-before-B3 and 300 B3-before-B2.
# Model phases avoid keeping two 26B GGUFs resident at once.
seed_indices = sorted({task['seed_index'] for task in selected_tasks})
for seed_index in seed_indices:
    seed_tasks = [task for task in selected_tasks if task['seed_index'] == seed_index]
    for ab_order in ['B2_then_B3', 'B3_then_B2']:
        block = [task for task in seed_tasks if task['ab_order'] == ab_order]
        if not block:
            continue
        phases = ['base', 'tuned'] if ab_order == 'B2_then_B3' else ['tuned', 'base']
        for phase in phases:
            phase_baselines = ['B1', 'B2'] if phase == 'base' else ['B3']
            pending = [(baseline, task) for task in block for baseline in phase_baselines if baseline in RUN_BASELINES and (baseline, task['task_id']) not in completed]
            if not pending:
                continue
            if phase == 'base':
                load_llm('base_gemma4_26b_q3km', BASE_MODEL)
                for task in block:
                    local_order = sorted(['B1', 'B2'], key=lambda b: task['baseline_order'].index(b))
                    for baseline in local_order:
                        run_one(baseline, task)
            else:
                load_llm('finetuned_gemma4_26b_q3km', TUNED_MODEL)
                for task in block:
                    run_one('B3', task)
            unload_llm()

print('Runner finished. Events for this signature:', len(completed), '/', len(selected_tasks) * len(RUN_BASELINES))
```
